```python
from keras.models import Model
from keras.layers import Dense, Input, BatchNormalization, Activation, Conv2D, MaxPooling2D, AveragePooling2D, Dropout
from keras.layers import Add, Reshape, Flatten
from keras import backend as K
from keras.utils import plot_model
from keras.utils.vis_utils import model_to_dot
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from ImageUtils import ImageUtils
```

```python
'''构建残差网络的identity unit, and add Dropout operation to it'''
def identity_unit(X, filters = 64, kernel_size = (3, 3), strides = 1, padding = 'same', dropout = 0.5):
    Y = Conv2D(filters, kernel_size, strides = strides, padding = padding)(X)
    Y = Activation('relu')(Y)
    Y = Dropout(dropout)(Y)
    Y = Conv2D(filters, kernel_size, strides = strides, padding = padding)(Y)
    X = Conv2D(filters, kernel_size, strides = strides, padding = padding)(X)
    Y = Add()([X, Y])
    Y = Activation('relu')(Y)
    Y = Dropout(dropout)(Y)
    return Y
```

```python
#测试identity_unit 方法
X1 = Input(shape = (320, 320, 3))
X_test1 = np.random.rand(1, 320, 320, 3)
output = identity_unit(X1)
model_test1 = Model(inputs = X1, outputs = output )
Y1 = model_test1.predict(X_test1)
print(Y1.shape)
```

```python
'''bottleneck block of residual unit'''
def bottleneck_unit(X, filters = (64, 256), kernel_size = ((1, 1), (3, 3)), strides = 1, padding = 'same', dropout = 0.5):
    Y = Conv2D(filters[0], kernel_size[0], strides = strides, padding = padding)(X)
    X = Conv2D(filters[1], kernel_size[0], strides = strides, padding = padding)(X)
    Y = Activation('relu')(Y)
    Y = Conv2D(filters[0], kernel_size[1], strides = strides, padding = padding)(Y)
    Y = Activation('relu')(Y)
    Y = Conv2D(filters[1], kernel_size[1], strides = strides, padding = padding)(Y)
    Y = Add()([Y, X])
    Y = Activation('relu')(Y)
    Y = Dropout(dropout)(Y)
    return Y
```

```python
'''测试 bottleneck_unit'''
inputs = Input(shape = (320, 320, 3))
data = np.random.rand(1, 320, 320, 3)
outputs = bottleneck_unit(inputs)
model_test2 = Model(inputs = inputs, outputs = outputs)
Y2 = model_test2.predict(data)
print(Y2.shape)
```

```python
'''构建深层resnet'''
def resnet(input_shape, reshape, layers = 5):
    inputs = Input(shape = input_shape)
    X = Reshape(reshape, input_shape = input_shape)(inputs)
    X = Conv2D(64, (7, 7), strides = 2, padding = 'valid')(X)
    for i in range(layers):
        X = identity_unit(X)
    for i in range(layers):
        X = bottleneck_unit(X)
    X = AveragePooling2D(pool_size = (2, 2), strides = 1, padding = 'valid')(X)
    X = Activation('relu')(X)
    X = Flatten()(X)
    outputs = Dense(7, activation = 'relu')(X)
    model = Model(inputs = inputs, outputs = outputs)
    return model
```

```python
'''build model obejct'''
model = resnet(input_shape = (320, 320), reshape = (320, 320, 1))
```

```python
'''test model'''
inputs = np.random.rand(5, 320, 320)
model.predict(inputs)
```

```python
'''compile model'''
model.compile(loss = 'mean_squared_error', optimizer = 'adadelta', metrics = ['accuracy'])
```

```python
'''train model'''
model.fit(X_train, Y_train, batch_size = 32, epochs = 5)
```
