```python
# -*- coding:utf-8 -*-
from bs4 import BeautifulSoup
import urllib
from urllib.request import Request, urlopen, urlretrieve
import http
import time
import sys
import os
import json
```

```python
data_root = '.'
class crawler:
    def __init__ (self, baseUrl, name, page, rn = 30, pn = 0, sleep = 0.1):
        self.name = urllib.parse.quote(name)
        self.baseUrl = baseUrl
        self.rn = rn
        self.pn = pn
        self.sleep = sleep
        self.count = 0
        self.page = page
    def getUrl (self):
        return self.baseUrl + ''.join(['&', 'queryWord', '=', self.name, '&', 'word', '=', self.name, '&', 'pn', '=', str(self.pn), '&', 'rn', '=', str(self.rn)])
    def downJson (self):
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:23.0) Gecko/20100101 Firefox/23.0'}
        req = Request(self.getUrl(), headers =  headers)
        res = urlopen(req)
        return res.read().decode('utf8')
    def parseUrlFromJson (self):
        urls = []
        try:
            json_data = json.loads(self.downJson())['data']
            for i in range(len(json_data)):
                replaceUrl = json_data[i]['replaceUrl']
                for j in range(len(replaceUrl)):
                    urls.append(replaceUrl[j]['ObjURL'])
        except Exception as e:
            return urls
    def downImag(self):
        time.sleep(self.sleep)
        urls = self.parseUrlFromJson()
        for url in urls:
            try:
                subfix = url.split('.').pop()
                urlretrieve(url, os.path.join(data_root, 'img/' + str(self.count) + '.' + subfix ))
                self.count = self.count + 1
            except Exception as e:
                print(e)
            
    def start(self):
        for i in range(self.page):
            self.pn = (i - 1) * self.rn
            self.downImag()

query = {
    'baseUrl': 'http://image.baidu.com/search/acjson?tn=resultjson_com&ipn=rj&ct=201326592&is=&fp=result&cl=2&lm=-1&ie=utf-8&oe=utf-8&adpicid=&st=&z=&ic=&s=&se=&tab=&width=&height=&face=&istype=&qc=&nc=&fr=',
    'pn': 0,
    'rn': 30,
    'sleep': 1,
    'name': '车牌',
    'page': 100
}

c = crawler(query['baseUrl'], query['name'], query['page'], query['rn'], query['pn'], query['sleep'])
c.start()
```
